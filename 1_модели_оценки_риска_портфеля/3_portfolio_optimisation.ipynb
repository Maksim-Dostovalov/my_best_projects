```python
def create_random_weights(n_assets):
    """
    Возвращает случайно выбранные веса портфеля, которые в сумме равны единице
    """

    w = np.random.rand(n_assets)
    return w / w.sum()

def evaluate_random_portfolio(returns):
    """
    Возвращает среднее и стандартное отклонение доходности для случайного портфеля
    """

    returns = pd.DataFrame(returns)

    # рассчитать на основе ковариации доходности активов и весов
    cov = np.matrix(returns.cov())
    R = np.matrix(returns.mean())
    w = np.matrix(create_random_weights(returns.shape[1]))

    # рассчитать ожидаемую доходность и риск портфеля
    mu = w * R.T
    sigma = np.sqrt(w * cov * w.T) # Среднеквадратическое отклонение

    return mu, sigma

def create_random_portfolios(returns, n_portfolios=1500):
    '''
    Созданные случайным образом портфели
    '''
    # рассчитать среднее и среднее значение для каждого портфеля
    pf_mus, pf_sigmas = np.column_stack([evaluate_random_portfolio(returns) for _ in range(n_portfolios)])

    return pf_mus, pf_sigmas
```

```python
import cvxopt as opt
from cvxopt import solvers

def calculate_frontier(returns):
    '''
    возвращает оптимальные веса портфеля и соответствующие сигмы для желаемой оптимальной доходности портфеля
    Params:
    - returns: T x N matrix of observed data
    '''
    returns = pd.DataFrame(returns)
    cov = np.matrix(np.cov(returns.T))
    N = returns.shape[1]
    pbar = np.matrix(returns.mean())

    # определить список оптимальных / желаемых mus, для которых мы хотели бы найти оптимальные сигмы
    optimal_mus = []
    r_min = pbar.mean()    # минимальная ожидаемая доходность
    for i in range(50):
        optimal_mus.append(r_min)
        r_min += (pbar.mean() / 80)

    # матрицы ограничений для квадратичного программирования
    P = opt.matrix(cov)
    q = opt.matrix(np.zeros((N, 1)))
    G = opt.matrix(np.concatenate((-np.array(pbar), -np.identity(N)), 0))
    A = opt.matrix(1.0, (1,N))
    b = opt.matrix(1.0)

    # оптимизация скрытия
    opt.solvers.options['show_progress'] = False

    # вычислить веса портфеля, каждый вектор веса имеет размер Nx1
    # найти оптимальные веса с помощью qp(P, q, G, h, A, b)
    optimal_weights = [solvers.qp(P, q, G, opt.matrix(np.concatenate((-np.ones((1, 1)) * mu, np.zeros((N, 1))), 0)), A, b)['x'] for mu in optimal_mus]

    # найти оптимальную сигму
    # \sigma = w^T * Cov * w
    optimal_sigmas = [np.sqrt(np.matrix(w).T * cov.T.dot(np.matrix(w)))[0,0] for w in optimal_weights]

    return optimal_weights, optimal_mus, optimal_sigmas
```

# Симуляция Монте-Карло для оптимизированного портфеля

```python
def monte_carlo(tickers, weights, T, mc_sims):  # T - количество недель, mc_sims - количество симуляций

    # Ковариационная матрица
    covMatrix = tickers.cov()

    # Средняя доходность
    meanReturns = tickers.mean()

    meanM = np.full(shape=(T, len(weights)), fill_value=meanReturns)
    meanM = meanM.T

    portfolio_sims = np.full(shape=(T, mc_sims), fill_value=0.0)

    for m in range(0, mc_sims):
        Z = np.random.normal(size=(T, len(weights))) # Генерация случайных значений для каждой акции в портфеле на каждую неделю моделирования
        L = np.linalg.cholesky(covMatrix) # Разложение Холецкого ковариационной матрицы для получения нижнетреугольной матрицы
        dailyReturns = meanM + np.inner(L, Z) # Коррелированные дневные доходы по отдельным акциям
        portfolio_sims[:,m] = np.cumprod(np.inner(weights, dailyReturns.T)+1)-1

    return portfolio_sims
```

```python
def value_at_risk(returns, confidence_level):
    """
    It calculates the Value at Risk (VaR) of some time series. It represents
    the maximum loss with the given confidence level.

    Parameters
    ----------
    returns : pandas.DataFrame
        Returns of each time series. It could be daily, weekly, monthly, ...

    confidence_level : int
        Confidence level. 5% by default.

    Returns
    -------
    var : pandas.Series
        Value at Risk for each time series.

    """

    # Calculating VaR
    return returns.quantile(confidence_level, interpolation='higher')
```

```python
def expected_shortfall(returns, confidence_level):
    """
    It calculates the Expected Shortfall (ES) of some time series. It represents
    the average loss according to the Value at Risk.

    Parameters
    ----------
    returns : pandas.DataFrame
        Returns of each time series. It could be daily, weekly, monthly, ...

    confidence_level : int
        Confidence level. 5% by default.

    Returns
    -------
    es : pandas.Series
        Expected Shortfall for each time series.

    """

    # Calculating VaR
    var = value_at_risk(returns, confidence_level)

    # ES is the average of the worst losses (under var)
    return returns[returns.lt(var)].mean()
```
